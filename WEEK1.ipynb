```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('SupplyChainEmissionFactorsforUSIndustriesCommodities(2016_Summary_Commodity).csv')
df.head()
```

```python
print(df.shape)
```

```python
plt.scatter(df['DQ ReliabilityScore of Factors without Margins'], df['Supply Chain Emission Factors without Margins'])
plt.xlabel('DQ ReliabilityScore of Factors without Margins')
plt.ylabel('Supply Chain Emission Factors without Margins')
plt.title('Scatter plot of Emission Factors vs Reliability Score')
plt.show()
```

```python
df_sorted = df.sort_values('Supply Chain Emission Factors with Margins', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['Commodity Name'], df_sorted['Supply Chain Emission Factors with Margins'])
plt.xlabel('Commodity Name')
plt.ylabel('Supply Chain Emission Factors with Margins')
plt.title('Top 10 Commodities by Supply Chain Emission Factors')
plt.xticks(rotation=80) 
plt.show()
```
